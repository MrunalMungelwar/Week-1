This notebook explores air quality data from Indian cities between 2015 and 2020.

```python
# 1. Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

```python
# 2. Load the dataset
# (Download 'city_day.csv' from Kaggle Air Quality dataset and keep in same folder)
df = pd.read_csv("city_day.csv")
```

```python
# 3. Explore the dataset
print("Shape of Dataset:", df.shape)
print("\nFirst 5 Rows:\n", df.head())

print("\n--- Dataset Info ---")
print(df.info())

print("\n--- Statistical Summary ---")
print(df.describe())

print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python
# Optional: Check AQI bucket distribution if available
if 'AQI_Bucket' in df.columns:
    print("\nAQI_Bucket Value Counts:\n", df['AQI_Bucket'].value_counts())

    # Plot AQI category distribution
    plt.figure(figsize=(7,4))
    sns.countplot(data=df, x="AQI_Bucket", palette="coolwarm", order=df["AQI_Bucket"].value_counts().index)
    plt.title("Distribution of AQI Categories")
    plt.xticks(rotation=30)
    plt.show()
```

```python
# Optional: Correlation Heatmap for pollutants
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlOrRd", fmt=".2f")
plt.title("Correlation Heatmap of Pollutants & AQI")
plt.show()
```
